When examining an $AR(1)$ process, we refer to this process as a **Random Walk Process** if $\alpha$ in the expression $ y_t = \alpha \cdot y_{t-1} + u_t $ (Equation 1) is equal to 1. In this case, the process is non-stationary, implying that it contains a unit root. For this non-stationary process, we set up the following hypotheses: $ H_0:\alpha=1 $ and $ H_1:\alpha<1 $. The null hypothesis states that the series $ y_t $ contains a unit root, indicating that it is non-stationary. For an $AR(1)$ process to be stationary, the roots of the equation must have an absolute value less than one. Conversely, the alternative hypothesis $ H_1 $ assumes that $ \alpha $ is less than 1, suggesting that the series $ y_t $ does not contain a unit root (i.e., it is stationary). Now, letâ€™s explore the process in Equation 1. By subtracting $ y_{t-1} $ from both sides, we get:

$$
y_t - y_{t-1} = \alpha \cdot y_{t-1} - y_{t-1} + u_t
$$

Rearranging this expression yields:

$$
\Delta y_t = (\alpha - 1) y_{t-1} + u_t
$$

If we define this as Equation 2, we see that while Equation 1 and Equation 2 are mathematically equivalent, they represent the same process. Let us now denote $(\alpha - 1)$ as $\gamma$. This redefinition allows us to restate the hypotheses as follows: if $\alpha - 1 = \gamma$, the new hypotheses are $ H_0:\gamma=0 $ and $ H_1:\gamma<0 $. In this case, the null hypothesis implies that the series $ y_t $ contains a unit root, while the alternative hypothesis suggests that $ y_t $ is stationary. Notice that Equation 2 is in its most basic form, devoid of any deterministic components, meaning that it contains neither an intercept nor a trend. Letâ€™s designate Equation 2 as Model 1 and proceed to explore Models 2 and 3:

$$
\Delta y_t = \beta_0 + \gamma y_{t-1} + u_t
$$

We refer to this as Model 2, which includes an intercept. The third model is defined as:

$$
\Delta y_t = \beta_0 + \beta_1 \cdot t + \gamma \cdot y_{t-1} + u_t
$$

This model includes a trend term (denoted by $t$) and is called Model 3. Despite these model distinctions, it is crucial to note that the hypotheses remain consistent across all three models. Specifically, we are testing the hypothesis $ H_0:\gamma=0 $. To do so, we calculate the $t$-statistic, denoted as $ t_{\hat{\gamma}} $, which will be estimated using the Ordinary Least Squares (OLS) method. After estimating the models via OLS, we compute the respective $t$-statistics as follows: $ t_{\hat{\gamma}} = \frac{\hat{\gamma}}{s(\hat{\gamma})} $. These calculated statistics are then compared to critical values to determine the validity of the hypotheses. In standard regression models, such as those based on cross-sectional data, the $t$-statistic is assumed to follow a standard distribution. However, under the null hypothesis that the series is non-stationary, the $t_{\hat{\gamma}}$ statistic does not follow a standard Student's $t$ distribution. Instead, it has a right-skewed distribution. To address this issue, Dickey and Fuller (Dickey was a student of Fuller and developed this approach as part of his 1976 thesis, later published in a 1979 paper) used Monte Carlo simulation methods to derive critical values. This critical table helps to make a decision regarding the null hypothesis. This is the first step in the process: once the model is estimated at its levels, a decision is made on whether to reject the null hypothesis. If the null hypothesis is rejected in Model 1, it implies that the $ y_t $ series is stationary. If not, it suggests that the original series contains a unit root. Determining the stationarity of the $ y_t $ series prior to analysis is essential. Here are the models once again:

$$
\Delta y_t = \gamma \cdot y_{t-1} + \ldots + u_t
$$

$$
\Delta y_t = \beta_0 + \gamma \cdot y_{t-1} + \ldots + u_t
$$

$$
\Delta y_t = \beta_0 + \beta_1 \cdot t + \gamma \cdot y_{t-1} + \ldots + u_t
$$

The models are referred to as Model 1, Model 2, and Model 3, respectively. According to Dickey and Fuller (1981), if the residuals $ u_t $ exhibit autocorrelation, one of the simplest ways to address this is by including lagged values of the dependent variable in the model. Let us therefore add the lags of the dependent variable $ y_t $ to the model, represented as:

$$
\Delta y_t = \gamma \cdot y_{t-1} + \sum_{j=1}^{k} \delta_j \Delta y_{t-j} + u_t
$$

$$
\Delta y_t = \beta_0 + \gamma \cdot y_{t-1} + \sum_{j=1}^{k} \delta_j \Delta y_{t-j} + u_t
$$

$$
\Delta y_t = \beta_0 + \beta_1 \cdot t + \gamma \cdot y_{t-1} + \sum_{j=1}^{k} \delta_j \Delta y_{t-j} + u_t
$$

Here, we test the hypothesis $ H_0:\gamma=0 $, where $k$ is selected based on model selection criteria, such as the AIC or SIC. Once a criterion is chosen, it is advisable to stick to it consistently to avoid criticism. Although model selection criteria do not alter the hypothesis being tested, it is preferable to remain consistent in the selection method. Now, letâ€™s estimate Model 3 using OLS and test the significance of the $ t $ (trend) coefficient. If it is significant, we test $ H_0:\gamma=0 $. Rejecting $ H_0 $ indicates that the $ y_t $ series is stationary. If $ H_0 $ cannot be rejected, we take the first difference of $ y_t $, denoted by $ \Delta y_t $. For the differenced series $ \Delta y_t $, it is unnecessary to estimate the trend model. Removing the trend results in the model being trend-free. Briefly, the coefficients $ \beta_0 $ and $ \beta_1 $ in this equation are assumed to follow a $t$ distribution. Next, Model 2 is estimated, and $ H_0:\gamma=0 $ is tested. If $ H_0 $ is rejected, we conclude that $ \Delta y_t $ is stationary, and consequently, $ y_t $ is first-order differenced stationary. Thus, the original series becomes stationary at the first difference, denoted by $ y_t \sim I(1) $. However, if $ H_0 $ cannot be rejected, we take the second difference of $ y_t $, denoted by $ \Delta^2 y_t $, and estimate Model 2 once more. Testing $ H_0:\gamma=0 $ in this context, if $ H_0 $ is rejected, it implies that $ \Delta^2 y_t $ is stationary, and the process is $ y_t \sim I(2) $. This scenario, however, is relatively rare in macroeconomic series. If the trend in Model 3 is found to be insignificant, we proceed with Model 2 for the $ y_t $ series, testing $ H_0:\gamma=0 $. Rejecting $ H_0 $ implies that $ y_t \sim I(0) $; if not, we repeat the test for $ \Delta y_t $ in Model 2. Upon rejection of this hypothesis, $ \Delta y_t $ is considered $ I(0) $, whereas the original series is taken as $ I(1) $. As mentioned at the beginning of the project, the integration of unit root tests for each variable can be formulated as follows using three different model specifications: Model 1 (no constant, no trend), Model 2 (constant only), and Model 3 (constant and trend). Here are the mathematical definitions of each model for the variables **credit**, **usd**, **tufe**, and **ito**.

## 1. '**credit**'

- **Model 1** (No constant, no trend):
  $$
  \text{credit}_t = \gamma \cdot \text{credit}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{credit}_{t-j} + u_t
  $$

- **Model 2** (Constant only):
  $$
  \text{credit}_t = a_0 + \gamma \cdot \text{credit}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{credit}_{t-j} + u_t
  $$

- **Model 3** (Constant and trend):
  $$
  \text{credit}_t = a_0 + a_1 \cdot t + \gamma \cdot \text{credit}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{credit}_{t-j} + u_t
  $$

## 2. '**usd**'
- **Model 1** (No constant, no trend):
  $$
  \text{usd}_t = \gamma \cdot \text{usd}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{usd}_{t-j} + v_t
  $$

- **Model 2** (Constant only):
  $$
  \text{usd}_t = b_0 + \gamma \cdot \text{usd}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{usd}_{t-j} + v_t
  $$

- **Model 3** (Constant and trend):
  $$
  \text{usd}_t = b_0 + b_1 \cdot t + \gamma \cdot \text{usd}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{usd}_{t-j} + v_t
  $$

## 3. '**tufe**'

- **Model 1** (No constant, no trend):
  $$
  \text{tufe}_t = \gamma \cdot \text{tufe}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{tufe}_{t-j} + \eta_t
  $$

- **Model 2** (Constant only):
  $$
  \text{tufe}_t = \alpha_0 + \gamma \cdot \text{tufe}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{tufe}_{t-j} + \eta_t
  $$

- **Model 3** (Constant and trend):
  $$
  \text{tufe}_t = \alpha_0 + \alpha_1 \cdot t + \gamma \cdot \text{tufe}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{tufe}_{t-j} + \eta_t
  $$

Explanation of Parameters;

- $ \Delta y_t $: The first difference of the variable $ y $ at time $ t $.
- $ \gamma $: The coefficient testing for unit root (if $ \gamma = 0 $, the series is non-stationary).
- $ a_0, b_0, \alpha_0 $: The constant term in Model 2 (for each respective variable).
- $ a_1, b_1, \alpha_1 $: The coefficient of the trend term $ t $ in Model 3 (for each respective variable).
- $ \sum_{j=1}^{k} \delta_j \Delta y_{t-j} $: The lagged differences of the variable $ y $, included to account for autocorrelation.
- $ u_t, v_t, \eta_t $: The error term.

Each model tests for the presence of a unit root in the series by examining the coefficient $ \gamma $. The inclusion of a constant and/or trend term depends on the assumptions about the series behavior. In this context, the objective is to perform unit root tests and assess the stationarity structures of each series within different sections. Specifically, **SECTION 1** focuses on the `credit` series, **SECTION 2** on the `usd` series, **SECTION 3** on the `tufe` series. The analysis aims to test the stationarity of these series and examine them visually, providing insights into their underlying patterns and stability over time.

Upon examining the correlograms, the ACF graph reveals a noticeable decline in the series. To further investigate, let's take the first difference of the 'credit' series and recreate the correlograms.

```python

credit_diff = ADFdataset['credit'].astype(float).diff().dropna()
# ACF ve PACF Graph size;
plt.figure(figsize=(12, 4))

# ACF;
plt.subplot(1, 2, 1)
plot_acf(credit_diff, 
         lags=36, 
         ax=plt.gca(), 
         title="ACF - First Difference of 'credit' Series")
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")

# PACF;
plt.subplot(1, 2, 2)
plot_pacf(credit_diff, 
          lags=36, 
          ax=plt.gca(), 
          title="PACF - First Difference of 'credit' Series", 
          method="ywm")
plt.xlabel("Lag")
plt.ylabel("Partial Autocorrelation")
plt.tight_layout()
plt.show()
```

When examining the ACF values of the first-differenced 'credit' variable, namely the 'credit_diff' variable, we observe that after the fourth lag, the ACF values generally trend towards zero and remain within the confidence bands (with a $2/\sqrt{n}$). Comparing this with the original 'credit' series, we can clearly see that the slowly decaying structure of the original series has been disrupted. This preliminary analysis suggests that the first difference of the series may indeed be stationary.

```python

plt.figure(figsize=(10, 4))
plt.plot(ADFdataset['date'].iloc[1:], 
         credit_diff, 
         color="navy", 
         linewidth=1.8, 
         linestyle="-")
plt.title("First Difference of Debit Card and Credit Card Spending (Current Data, Thousand TL)", 
          fontsize=12, 
          weight='bold')
plt.xlabel("Date", 
           fontsize=12)
plt.ylabel("First Difference of Credit (Value)", 
           fontsize=12)
plt.xticks(rotation=45)
plt.grid(visible=True, 
         linestyle="--", 
         alpha=0.7)
plt.tight_layout()
plt.show()
```

Although certain distortions are observed in the first-differenced `credit` series within the 2022-2024 period, it appears to have become stationary. Let's now proceed with the stationarity analysis by conducting the Augmented Dickey-Fuller test on the 'credit' series. The hypotheses being tested were defined at the beginning of the project.

```python

# Let us estimate the Augmented Dickey-Fuller test (including the trend and intercept) with ECM using maxlag=14; 
# AIC information criterion and analyze the stationarity results for the 'credit' series;
adf_result = adfuller(credit, 
                      maxlag=14, 
                      regression='ct', 
                      autolag='AIC')

# Let's print the test results;
print("ADF Statistic:", 
      adf_result[0])
print("p-value:", 
      adf_result[1])
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")

# OLS regression to estimate trend and constant term coefficients;
# Let's create the trend term by adding a time variable;
credit = pd.Series(credit).dropna().reset_index(drop=True)
time_trend = range(len(credit)) # Trend

# Let's build the data frame using the constant term and trend;
X = pd.DataFrame({'const': 1, 
                  'trend': time_trend})
ols_model = sm.OLS(credit, X).fit()  # Let's estimate the regression with OLS.

# Let's print the coefficients of the trend and the constant term;
print("\nOLS Regression Results:")
print(ols_model.summary())
```

$$\Delta \text{credit}_t = a_0 + a_1 \cdot t + \gamma \cdot \text{credit}_{t-1} + \sum_{j=1}^{k} \delta_j \Delta \text{credit}_{t-j} + u_t$$
When examining the initial results of `Model 3`, we find that the ADF test statistic is $-0.7617$. Before comparing the series' stationarity hypothesis with critical values, we observe that the ADF statistic, in absolute terms, is above the critical values at the $1\%$, $5\%$, and $10\%$ levels. This suggests that the series is not stationary. Additionally, with a high p-value, we could not reject the unit root hypothesis. Typically, when p < 0.05, we accept the series as stationary. At this point, it is evident that the `credit` series contains a unit root and is therefore non-stationary.

Upon reviewing the OLS regression results, our analysis based on the significance of the trend and intercept terms shows that the $trend$ coefficient is statistically significant, indicating a strong positive trend in the `credit` series. The positive coefficient reflects an increase in the `credit` series over time. At this stage, a visual examination of the `credit` series, combined with the unit root test results, shows that the series exhibits a non-stationary structure.

```python

# ADF Test;
adf_result = adfuller(credit_diff, 
                      maxlag=14, 
                      regression='ct', 
                      autolag='AIC')

# Test results;
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")

# OLS regression with constant term and trend;
credit_diff = pd.Series(credit_diff).dropna().reset_index(drop=True)

# Create data frame with constant term and trend;
time_trend = range(len(credit_diff))
X = pd.DataFrame({'const': 1, 
                  'trend': time_trend}, 
                 index=credit_diff.index)
ols_model = sm.OLS(credit_diff, X).fit()

# Let's print the constant term and trend coefficients;
print("\nOLS Regression Results:")
print(ols_model.summary())
```

When the ADF critical values of the first order differenced `credit` series are analyzed, and the $p$-value is greater than all three probability values in absolute terms, the null hypothesis is rejected and the series still contains a unit root. It is observed that the series with first order difference ($\Delta credit_t$) is not stationary at both level and first order difference.

At this stage, let's take the second order difference of the series and subject it to the ADF test again;

```python

# Create a second row differenced credit series;
credit_diff2 = credit.diff().diff().dropna()

adf_result = adfuller(credit_diff2, 
                      maxlag=14, 
                      regression='ct', 
                      autolag='AIC')

# Let's print the test results;
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")

# OLS regression with constant term and trend;
credit_diff2 = pd.Series(credit_diff2).dropna().reset_index(drop=True)

# Creating a data frame with constant term and trend;
time_trend = range(len(credit_diff2))
X = pd.DataFrame({'const': 1, 
                  'trend': time_trend}, 
                 index=credit_diff2.index)
ols_model = sm.OLS(credit_diff2, X).fit()

# Printing constant term and trend coefficients;
print("\nOLS Regression Results:")
print(ols_model.summary())
```

Upon examining the results of the ADF tests conducted on the second-differenced `credit` series ($\Delta^2credit_t$), we observe that the second-differenced `credit` series has become stationary. At the 1% significance level, we find that this series does not contain a unit root and has a stable structure around the mean.

As a brief reminder; if $credit_t \sim I(d)$, then $\Delta^d credit_t \sim I(0)$
$$\Delta^2credit_t \sim I(0) \to credit_t \sim I(2)$$
$$\Delta credit_t \sim I(0) \to credit_t \sim I(1)$$
$$credit_t \sim I(0) \to credit_t \sim I(0)$$

Thus, the $credit_t$ series is non-stationary, and while the $\Delta credit_t$ series is also non-stationary, the $\Delta^2credit_t$ series is stationary. Therefore, the $credit_t$ series is $credit_t \sim I(2)$, meaning it becomes stationary at the second difference. The `credit` series can also be analyzed with a logarithmic transformation or logarithmic differencing prior to unit root testing.

```python

# ADF test (including trend and constant term) for the 'usd' series;
adf_result_usd = adfuller(usd, 
                          maxlag=14, 
                          regression='ct', 
                          autolag='AIC')

# Print the ADF test results;
print("ADF Statistic:", adf_result_usd[0])
print("p-value:", adf_result_usd[1])
print("Critical Values:")
for key, value in adf_result_usd[4].items():
    print(f"   {key}: {value}")

# Let's estimate the trend and constant term coefficients with OLS regression;
# We create a time variable to add the trend variable;
usd_series = pd.Series(usd).dropna().reset_index(drop=True)
time_trend = range(len(usd_series))  # Trend

# Let's create the data frame containing the constant term and trend;
X_usd = pd.DataFrame({'const': 1, 
                      'trend': time_trend})
ols_model_usd = sm.OLS(usd_series, X_usd).fit()

# OLS regression results;
print("\nOLS Regression Results:")
print(ols_model_usd.summary())
```

Upon examining the test results, we observe that the ADF statistic is $-2.3484$ with a $p$-value of 0.4074. The ADF test indicates that the `usd` series falls below the critical values in absolute terms, suggesting that the series is non-stationary (i.e., it contains a unit root). Consequently, we fail to reject the null hypothesis established at the beginning of the project, which assumed the presence of a unit root in the `usd` series. Looking at the OLS results, the trend coefficient is $0.0390$, which is statistically significant. This positive trend implies that the `usd` series has shown an upward movement over time. Thus, the series does not fluctuate around a stable mean in the long term.

Then, let's take the first order difference of the series and subject it to the ADF test again;

```python

# Applying the ADF test on the first-differenced 'usd' series with trend and intercept;
usd_diff_series = pd.Series(usd).diff().dropna().reset_index(drop=True)

# Conducting the ADF test (including trend and intercept) for the usd_diff series;
adf_result_usd_diff = adfuller(usd_diff_series, 
                               maxlag=14, 
                               regression='ct', 
                               autolag='AIC')

# Printing the ADF test results;
print("ADF Statistic:", adf_result_usd_diff[0])
print("p-value:", adf_result_usd_diff[1])
print("Critical Values:")
for key, value in adf_result_usd_diff[4].items():
    print(f"   {key}: {value}")

# Preparing data for OLS regression including intercept and trend;
time_trend = range(len(usd_diff_series))
X_usd_diff = pd.DataFrame({'const': 1, 
                           'trend': time_trend}, 
                          index=usd_diff_series.index)

# Estimating OLS regression with intercept and trend;
ols_model_usd_diff = sm.OLS(usd_diff_series, X_usd_diff).fit()

# Printing the OLS regression results;
print("\nOLS Regression Results:")
print(ols_model_usd_diff.summary())
```

Upon examining the results, we observe that the $\Delta usd_t$ series is strongly stationary. The critical values at each significance level are exceeded in absolute terms by the ADF statistic, confirming the absence of a unit root in the differenced series. Thus, the series has a stable structure around a long-term mean. The trend coefficient is statistically insignificant, indicating that no deterministic trend remains in the $\Delta usd_t$ series. This implies that the series fluctuates around a constant mean over time without any underlying upward or downward trend, highlighting its stationary nature. As a reminder of the integration structure:
If $usd_t \sim I(d)$, then $\Delta^d usd_t \sim I(0)$.
In this case:

$$
\Delta usd_t \sim I(0) \to usd_t \sim I(1)
$$

Therefore, the original $usd_t$ series is integrated of order 1, or $usd_t \sim I(1)$. This suggests that the `usd` series becomes stationary upon first differencing, indicating it has a stochastic trend at level. ```python

```python
# Assuming 'ADFdataset' is available and contains the 'tufe' series;
# Perform the ADF test on the 'tufe' series with trend and intercept;
tufe_series = ADFdataset['tufe'].astype(float).dropna()
adf_result_tufe = adfuller(tufe_series, maxlag=14, regression='ct', autolag='AIC')

# Print ADF test results;
print("ADF Statistic:", adf_result_tufe[0])
print("p-value:", adf_result_tufe[1])
print("Critical Values:")
for key, value in adf_result_tufe[4].items():
    print(f"   {key}: {value}")

# Perform OLS regression to estimate trend and constant term coefficients;
# Create a time variable for the trend term;
tufe_series = tufe_series.reset_index(drop=True)
time_trend = range(len(tufe_series))

# Data frame with constant and trend;
X_tufe = pd.DataFrame({'const': 1, 'trend': time_trend})
ols_model_tufe = sm.OLS(tufe_series, X_tufe).fit()

# Print OLS regression results;
print("\nOLS Regression Results:")
print(ols_model_tufe.summary())
```

The ADF statistic (2.5474) is less than all critical values in absolute terms, indicating that the series likely contains a unit root and is thus non-stationary. The p-value of 1.0 further confirms that the `tufe` series is not stationary at its level; in other words, we fail to reject the null hypothesis that the series contains a unit root. This suggests that the series may follow a distinct trend over time rather than oscillating around a long-term mean. The OLS model reveals a significant positive trend, with the trend coefficient at 12.9182 and a p-value of 0.000, indicating that the `tufe` series has an upward trend over time. This trendâ€™s significance suggests that non-stationarity in the `tufe` series is largely due to this inherent trend component. Given these results, the `tufe` series is non-stationary, primarily due to the trend component. To achieve stationarity, it may be necessary to difference the series. Mathematically, we test the following hypothesis: 
$$ H_0: \gamma = 0 \quad (\text{unit root present, non-stationary}) $$

$$ H_1: \gamma < 0 \quad (\text{no unit root, stationary}) $$

As we cannot reject $ H_0 $, we proceed by differencing the `tufe` series to examine for stationarity. Let us continue with the unit root tests by taking the first order difference of the series;

```python

# Assuming 'ADFdataset' is available and contains the 'tufe' series;
# Take the first difference of 'tufe' series;
tufe_diff_series = pd.Series(ADFdataset['tufe'].astype(float)).diff().dropna().reset_index(drop=True)

# Conduct the ADF test (including trend and intercept) on the first-differenced tufe series;
adf_result_tufe_diff = adfuller(tufe_diff_series, 
                                maxlag=14, 
                                regression='ct', 
                                autolag='AIC')

# Print ADF test results;
print("ADF Statistic:", adf_result_tufe_diff[0])
print("p-value:", adf_result_tufe_diff[1])
print("Critical Values:")
for key, value in adf_result_tufe_diff[4].items():
    print(f"   {key}: {value}")

# Prepare data for OLS regression including intercept and trend;
time_trend = range(len(tufe_diff_series))
X_tufe_diff = pd.DataFrame({'const': 1, 
                            'trend': time_trend}, 
                           index=tufe_diff_series.index)

# Estimating OLS regression with intercept and trend;
ols_model_tufe_diff = sm.OLS(tufe_diff_series, X_tufe_diff).fit()

# Print OLS regression results;
print("\nOLS Regression Results:")
print(ols_model_tufe_diff.summary())
```

The results indicate that $\Delta tufe_t$ remains non-stationary. The ADF statistic is below all critical values in absolute terms, suggesting that the series contains a unit root and is thus non-stationary. At this point, we cannot reject the null hypothesis formulated for the `tufe` series, indicating that stationarity has not been achieved. The significance of the trend coefficient shows that the series still follows an upward trend, with a tendency to increase over time. Therefore, let us take the second difference of the $tufe_t$ series and repeat the tests to assess stationarity once more.

```python

# Assuming 'ADFdataset' is available and contains the 'tufe' series;
# Take the second difference of 'tufe' series
tufe_diff2_series = pd.Series(ADFdataset['tufe'].astype(float)).diff().diff().dropna().reset_index(drop=True)

# Conduct the ADF test (including trend and intercept) on the second-differenced tufe series;
adf_result_tufe_diff2 = adfuller(tufe_diff2_series, 
                                 maxlag=14, 
                                 regression='ct', 
                                 autolag='AIC')

# Print ADF test results;
print("ADF Statistic:", adf_result_tufe_diff2[0])
print("p-value:", adf_result_tufe_diff2[1])
print("Critical Values:")
for key, value in adf_result_tufe_diff2[4].items():
    print(f"   {key}: {value}")

# Prepare data for OLS regression including intercept and trend;
time_trend = range(len(tufe_diff2_series))
X_tufe_diff2 = pd.DataFrame({'const': 1, 'trend': time_trend}, index=tufe_diff2_series.index)

# Estimating OLS regression with intercept and trend;
ols_model_tufe_diff2 = sm.OLS(tufe_diff2_series, X_tufe_diff2).fit()

# Print OLS regression results;
print("\nOLS Regression Results:")
print(ols_model_tufe_diff2.summary())
```

The ADF test results for the second-differenced `tufe` series clearly indicate that the series has now become stationary and no longer contains a unit root. The ADF statistic (-4.9945) is below all critical values in absolute terms, leading us to reject the null hypothesis $ H_0: \gamma = 0 $, which posits that the series contains a unit root. This conclusion is further supported by the very low p-value of 0.0002, suggesting that the `tufe` series achieves stationarity in its second difference. Examining the OLS regression results, we observe that the trend coefficient is not statistically significant (p > 0.05), indicating that the series no longer follows a distinct long-term trend and has now stabilized around a constant mean. Based on these results, we conclude that the `tufe` series is stationary in its second difference. Mathematically, the integration order of the series can be expressed as follows:

- If $ tufe_t \sim I(d) $, then for stationarity, $ \Delta^d tufe_t \sim I(0) $ must hold.

In this case:
$$ \Delta^2 tufe_t \sim I(0) \to tufe_t \sim I(2) $$ 

This outcome shows that the `tufe` series becomes stationary in its second difference and that the original series follows an $ I(2) $ process. Thus, the `tufe` series should be analyzed as stationary after taking the second difference.
